```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
url = "https://docs.google.com/spreadsheets/d/1uSqBu4S3fF5gpn6vGY2_3IlqmFG4E-lge66bcgrmnAs/export?format=csv"
df = pd.read_csv(url)
df
# Calculate the unemployment rate
unemployed = df['Estimated Unemployment Rate (%)']
labour_force = df['Estimated Labour Participation Rate (%)']
unemployment_rate = (unemployed / labour_force) * 100

# Print the unemployment rate
print(unemployment_rate)
```
